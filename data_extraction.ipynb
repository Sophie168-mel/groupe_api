```python
df_commune = pd.read_csv("communes-departement-region.csv")
df_commune = df_commune[["code_commune_INSEE","nom_commune","code_departement","nom_departement","code_region","nom_region"]]
df_commune = df_commune.rename(columns={"code_commune_INSEE":"code_commune"})
df_commune.head()
```
